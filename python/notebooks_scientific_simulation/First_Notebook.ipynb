```python
import sympy as sp

x = sp.Function('x')
v,d,k,m,t = sp.symbols('v d k m t')

eqn = sp.Eq(x(t).diff(t, 2), (-k/m) * x(t).diff(t)**2)
```

```python
sp.dsolve(eqn, ics={x(t).diff(t).subs(t,0):v, x(0):d})
```

```python
sp.plot(sp.sin(sp.Symbol('x')))
```
