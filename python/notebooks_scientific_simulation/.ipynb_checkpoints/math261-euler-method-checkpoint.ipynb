```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
```

```python
def exact(t):
    return np.e**(-t)
```

```python
def diff(t, y):
    return 2*y - 3*(np.e**(-t))
```

```python
def euler(dydt, y_initial, t_initial, t_final, step):
    t = [t_initial]
    y = [y_initial]
    iteration_count = int((t_final-t_initial)/step) + 1
    
    for i in range(iteration_count):
        t_next = t_initial + (i+1)*step
        y_next = y[i] + dydt(t[i], y[i])*step
        
        t.append(t_next)
        y.append(y_next)
        
    return [t, y]
```

```python
t = np.arange(0, 10.5, 0.5)
y_exact = exact(t)

t1, y1 = euler(diff, 1, 0, 10, 1) # step size of 1
t2, y2 = euler(diff, 1, 0, 10, 0.5) # step size of 0.5
t3, y3 = euler(diff, 1, 0, 10, 0.25) # step size of 0.25
t4, y4 = euler(diff, 1, 0, 10, 0.01) # step size of 0.01
```

```python
plt.title("Exact vs Euler")
plt.xlabel("t")
plt.ylabel("y")
plt.xlim(-0.5, 10)
plt.ylim(-2, 1.5)
plt.grid(True)
plt.axis('equal')
plt.autoscale(False)
plt.plot(t, y_exact, color="green", label=f'exact')
plt.plot(t1, y1, color='crimson', marker='.', markersize=5, label=f'step size 1')
plt.plot(t2, y2, color='orange', marker='.', markersize=5, label=f'step size 0.5')
plt.plot(t3, y3, color='yellow', marker='.', markersize=5, label=f'step size 0.25')
plt.plot(t4, y4, color='black', label=f'step size 0.01')

plt.legend()
#plt.savefig('plot.pdf')
plt.show()
```
