```python
def getVals(xdata, ydata):
    n = len(xdata)
    b = 0
    xSum = 0
    ySum = 0
    xAvg = 0
    yAvg = 0
    xySum = 0
    xSqrSum = 0
    for i in range(n):
        xSum += xdata[i]
    for i in range(n):
        ySum += ydata[i]
    xAvg = xSum/n
    yAvg = ySum/n
    for i in range(n):
        xySum += xdata[i] * ydata[i]
    for i in range(n):
        xSqrSum += xdata[i] * xdata[i]
    b = (xySum - yAvg*xSum)/(xSqrSum - xAvg * xSum)
    a = yAvg - b*xAvg
    return [a, b]
```

```python
#Testing the function
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 200)
y = 3*x + 10-np.random.random(200)*20 #add random noise to y = 3x

a, b = getVals(x, y)

plt.title("Linear Regression")
plt.grid(True)
plt.plot(x, y, 'r.', label="Data")
plt.plot(x, 3*x, 'g', label="Actual Function (y = 3x)")
plt.plot(x, a + b*x, label="Linear Regression")
plt.legend()
plt.show()
```
