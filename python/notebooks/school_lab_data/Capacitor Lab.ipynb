```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
distance = np.array([2.0, 4.0, 5.6, 8.2, 10.0])
capdist = np.array([0.89, 0.44, 0.32, 0.22, 0.18])

plt.plot(distance, capdist, color='crimson', marker='o')
plt.grid(True)
plt.title('Capacitance (pF) vs Plate Separation (mm)')
plt.ylabel('Capacitance (pF)')
plt.xlabel('Plate Separation (mm)')
```

```python
area = np.array([100, 170, 240, 310, 400])
caparea = np.array([0.44, 0.75, 1.06, 1.37, 1.77])

plt.plot(area, caparea, color='crimson', marker='o')
plt.grid(True)
plt.title('Capacitance (pF) vs Plate Area (mm²)')
plt.ylabel('Capacitance (pF)')
plt.xlabel('Plate Area (mm²)')
```
