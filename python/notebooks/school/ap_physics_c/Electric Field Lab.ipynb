```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
distance = [29.9, 49.9, 73.4, 99.3, 148.2, 229.8, 290.5]
strength = [87.6, 69.9, 32.7, 15.5, 3.98, 3.35, 1.06]

plt.title('Electric Field Strength (V/m) vs. Distance (cm)')
plt.xlabel('Distance (cm)')
plt.ylabel('Electric Field Strength (V/m)')

plt.plot(distance, strength, 'rs', label='Experimental')
plt.plot(distance, strength, 'r')
plt.plot(np.linspace(29.9, 290.5, 70), (8.988)/((np.linspace(29.9, 290.5, 70)*0.01)**2),'g', label='Calculated')
plt.legend()
```

```python
charge = [1, 2, 3, 4]
strength = [9.57, 19.1, 28.7, 38.3]

plt.title('Electric Field Strength (V/m) vs. Charge (nC)')
plt.xlabel('Charge (nC)')
plt.ylabel('Electric Field Strength (V/m)')

plt.plot(charge, strength, 'rs')
plt.plot(charge, strength, 'r')
```
