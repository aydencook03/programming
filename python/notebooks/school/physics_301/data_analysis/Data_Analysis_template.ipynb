```python
import numpy as np #library for array operations and high-level math functions
import matplotlib.pyplot as plt #for making plots

data = np.genfromtxt('./mean_730d.csv', dtype='float', delimiter=',')
```

```python
sdata = np.copy(data)
sdata[:,1] = sdata[:,1] - np.mean(sdata[:,1])
sdata[:,1] = sdata[:,1]/np.std(sdata[:,1])

fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot()
ax.scatter(sdata[:,0],sdata[:,1])
```

```python
XArr = sdata[:,1]
tArr = sdata[:,0]

omega_low = 2*np.pi/(tArr[-1]-tArr[0])
omega_high = 2*np.pi/(tArr[3]-tArr[0])
```

```python
nf = 500
omegaArr = np.linspace(omega_low, omega_high, nf)
PArr = np.zeros(nf)
```

```python
for i in range(nf):
    omega = omegaArr[i]
    xc = np.sum( XArr*np.cos(omega*tArr) )
    xs = np.sum( XArr*np.sin(omega*tArr) )
    cc = np.sum( np.cos(omega*tArr) * np.cos(omega*tArr) )
    ss = np.sum( np.sin(omega*tArr) * np.sin(omega*tArr) )
    cs = np.sum( np.cos(omega*tArr) * np.sin(omega*tArr) )
    omega_tau = np.arctan(2*cs/(cc-ss))/2
    ct = np.cos(omega_tau)
    st = np.sin(omega_tau)
    PArr[i] = 0.5*( (ct*xc + st*xs)*(ct*xc + st*xs)/(ct*ct*cc + 2*ct*st*cs + st*st*ss)  +  (ct*xs - st*xc)*(ct*xs - st*xc)/(ct*ct*ss - 2*ct*st*cs + st*st*cc) )
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()
ax.plot(2*np.pi/omegaArr, PArr)
#ax.set_xlim(10000,35000)
#ax.set_ylim(0,0.25)
#ax.set_yscale('log')
```
