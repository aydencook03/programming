```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

plt.figure()
plt.subplot(211) #2cols, 1rows, subplot #1
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k', t2, np.exp(-t2), 'g--')

plt.subplot(212) #2cols, 1rows, subplot #2
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import sympy as sp

t = sp.Symbol('t')

sp.plot(sp.exp(-t) * sp.cos(2*sp.pi*t), sp.exp(-t), (t, 0, 5))
```
