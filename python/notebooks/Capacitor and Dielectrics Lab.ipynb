<h1 style="text-align:center">First Part: Energy Stored in the Capacitor vs Voltage</h1>

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize
%matplotlib inline
```

```python
V = np.array([0.000, .130, .306, .480, .611, .742, .917, 1.048, 1.310, 1.500])
Q = np.array([0, .92e-13, 2.17e-13, 3.40e-13, 4.33e-13, 5.26e-13, 6.50e-13, 7.42e-13, 9.26e-13, 10.62e-13])
U = np.array([0, 0.06e-13, 0.33e-13, 0.83e-13, 1.32e-13, 1.95e-13, 2.96e-13, 3.89e-13, 6.08e-13, 7.97e-13])

Area = 0.0004 #mÂ²
distance = 0.005 #m

E = V/distance
V2 = V**2
Q2 = Q**2
E2 = E**2
u = U/(Area*distance) # J/m^3
```

```python
def linreg(x, a, b):
    return a + b*x
```

```python
#a, b = scipy.optimize.curve_fit(linreg, V2, U)[0]
#print(a, b)

plt.title('Energy vs. VoltageÂ²')
plt.ylabel('Energy (J)')
plt.xlabel('VoltageÂ² (VÂ²)')
plt.grid(True)
plt.plot(V**2, U, 'r.-', label="3.08e-18 + (3.54e-13)VÂ²")
plt.legend()
plt.show()
```

```python
# if U = 1/2 * C * V**2
# then U = 1/2 *C*V**2 = a+bV**2
# C ~ 2*b
C = 2*b
print(C)
```

```python
#a, b = scipy.optimize.curve_fit(linreg, Q2, U)[0]
#print(a, b)

plt.title('Energy vs. ChargeÂ²')
plt.xlabel('ChargeÂ² (CÂ²)')
plt.ylabel('Energy (J)')
plt.grid(True)
plt.plot(Q**2, U, 'b.-', label="2.54e-13 + (1.0)QÂ²")
plt.legend()
plt.show()
```

```python
#a, b = scipy.optimize.curve_fit(linreg, E2, u)[0]
#print(a, b)

plt.title('Energy Density (u) vs. Electric-FieldÂ² (EÂ²)')
plt.xlabel('Electric-FieldÂ² (V/m)Â²')
plt.ylabel('Energy Density (J/mÂ³)')
plt.grid(True)
plt.plot(E2, u, 'g.-', label='1.53e-12 + (4.43e-12)EÂ²')
plt.legend()
plt.show()
```
